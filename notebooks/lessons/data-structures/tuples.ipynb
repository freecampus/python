### Debugging checkpoint 1.1

> **Debugging checkpoint**
A one-item tuple needs a comma: `single = (42,)`. Without the comma, `(42)` is just the number `42` in parentheses. If a tuple unpacking line fails, count the values on the right and the names on the left.
