## 2. Set Operations

Combine sets with union, intersection, difference, and symmetric difference.

> **Analogy**
Set operations are like comparing club rosters: who is in either club, both clubs, or
only one club?

### What this means

Set operations answer questions about overlap and difference.

### Example 2
Predict what will happen before you run the code.

```python
python = {"Ada", "Grace", "Linus"}
data = {"Grace", "Katherine"}
print(python | data)
print(python & data)
print(python - data)
```
