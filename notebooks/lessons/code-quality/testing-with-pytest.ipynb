## 2. pytest

## 3. Edge Cases

Choose tests for boundaries, empty inputs, and unusual but valid situations.

> **Analogy**
Edge cases are the corners of a tablecloth: if the corners are secure, the middle is
usually easier.

### What this means

An edge case is an input near a boundary where assumptions often break.

### Example 3
Predict what will happen before you run the code.

```python
def first(items):
    if not items:
        return None
    return items[0]

assert first([]) is None
```
