## 2. Selecting and Filtering Data

Choose columns, rows, and conditions from a DataFrame.

> **Analogy**
Filtering is like using a sieve: keep rows that match the condition and let others pass
away.

### What this means

Selection narrows a table to relevant columns or rows.

### Example 2
Predict what will happen before you run the code.

```python
import pandas as pd

df = pd.DataFrame({"name": ["Ada", "Grace"], "score": [95, 88]})
high = df[df["score"] >= 90]
print(high[["name", "score"]])
```
