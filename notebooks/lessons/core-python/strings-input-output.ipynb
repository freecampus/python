## 1. Strings are text values

A **string** is a value that represents text. In Python, the most common way to
create a string is to put quotation marks around the text.

````python
name = "Ada"
language = "Python"
message = "Hello"
````

The quotation marks tell Python: "treat this as text". The quotes are not part
of the text displayed to the user; they are the boundary markers in your code.

### Example 1
Before running this code, predict the four printed lines.

````python
word = "Python"
print(word)
print(word[0])
print(word[-1])
print(f"I am learning {word}.")
````

### Step-by-step explanation

1. `word = "Python"`
   - Python creates a string value containing the six characters `P`, `y`, `t`,
     `h`, `o`, and `n`.
   - The name `word` now refers to that string.
2. `print(word)`
   - Python looks up `word` and displays the whole string: `Python`.
3. `print(word[0])`
   - Python strings are ordered. The first character is at index `0`, not index
     `1`.
   - `word[0]` is `"P"`.
4. `print(word[-1])`
   - Negative indexes count from the end.
   - `word[-1]` is `"n"`, the last character.
5. `print(f"I am learning {word}.")`
   - The `f` before the opening quote makes an **f-string**.
   - Anything inside `{...}` is evaluated by Python.
   - `{word}` is replaced by the value of `word`, so the output is
     `I am learning Python.`

### Why indexes start at 0

Think of indexes as distance from the beginning. The first character is zero
steps away from the beginning, so its index is `0`.

```text
string:  P  y  t  h  o  n
index:   0  1  2  3  4  5
negative -6 -5 -4 -3 -2 -1
```

### Joining strings

You can join strings with `+`:

````python
first = "Ada"
last = "Lovelace"
full_name = first + " " + last
print(full_name)
````

The middle string, `" "`, is a space. Without it, the result would be
`"AdaLovelace"`.

For longer messages, prefer f-strings because they are easier to read:

````python
score = 95
print(f"Ada scored {score} points.")
````

### Common mistakes

> **Common mistake**
**Mistake 1: forgetting the `f` in an f-string.**

````python
name = "Ada"
print("Hello, {name}!")
````

This prints `Hello, {name}!` because Python treats `{name}` as ordinary text.
Fix it by adding `f` before the string:

````python
print(f"Hello, {name}!")
````

> **Common mistake**
**Mistake 2: using an index that does not exist.**

````python
word = "Python"
print(word[10])
````

`word` has indexes `0` through `5`. Index `10` is outside the string, so Python
raises `IndexError`.

## 2. Input and output

`print()` sends information out of a program. `input()` brings information in
from the user.

````python
name = input("What is your name? ")
print(f"Hello, {name}!")
````

When Python runs the first line, it pauses and waits. Whatever the user types is
returned as a string and assigned to `name`.

### Example 2
Try this in Google Colab or a local notebook. In Colab, use the play button on
the left side of the code cell.

````python
age_text = input("How old are you? ")
age = int(age_text)
next_year = age + 1
print(f"Next year you will be {next_year}.")
````

### Step-by-step explanation

1. `input("How old are you? ")`
   - The prompt text is displayed.
   - The user types something, such as `12`.
   - Python returns the text value `"12"`.
2. `age_text = input(...)`
   - The name `age_text` stores that text.
3. `age = int(age_text)`
   - `int(...)` tries to convert the text into an integer.
   - If `age_text` is `"12"`, the result is the number `12`.
4. `next_year = age + 1`
   - Now the value is numeric, so arithmetic works.
5. `print(f"Next year you will be {next_year}.")`
   - The result is placed inside the message and printed.

### Why conversion matters

`input()` always returns a string. This surprises many beginners.

````python
number = input("Type a number: ")
print(number + number)
````

If the user types `5`, the output is `55`, not `10`, because Python is joining
strings. Convert first when you need arithmetic:

````python
number_text = input("Type a number: ")
number = int(number_text)
print(number + number)
````

### Common mistakes

> **Common mistake**
**Mistake: trying to convert text that is not a whole number.**

````python
age = int("twelve")
````

This raises `ValueError` because Python does not know how to turn the word
`"twelve"` into an integer. When using `input()`, assume users may type
unexpected text and be ready to validate it later.
