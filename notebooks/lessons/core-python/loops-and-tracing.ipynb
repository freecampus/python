## The problem this lesson solves

Computers are especially good at repetition. If you can write instructions for
one item, a loop can repeat those instructions for many items. Loops help you
print every name in a list, add many numbers, search for a match, or keep asking
until valid input appears.

The hard part for beginners is not typing `for` or `while`; it is tracking what
changes each time the loop repeats. This lesson teaches loops together with
**tracing**, the habit of writing down values as the program runs.

## How loops repeat work

A loop is a repeated path. The important debugging question is: what changes each
time around the loop?

## 1. `for` loops repeat once per item

A `for` loop is best when you already have a collection of items and want to do
something with each item.

````python
for item in collection:
    do_something_with(item)
````

Read this as: "for each item in this collection, run the indented block."

### Example 1
Predict the output before running the code.

````python
names = ["Ada", "Grace", "Katherine"]

for name in names:
    print(f"Hello, {name}")
````

### Step-by-step explanation

1. `names = ["Ada", "Grace", "Katherine"]`
   - Python creates a list of three strings.
2. `for name in names:`
   - Python starts the loop.
   - On the first trip, `name` refers to `"Ada"`.
3. `print(f"Hello, {name}")`
   - The indented line is the loop body.
   - It prints `Hello, Ada` on the first trip.
4. Python goes back to the top of the loop.
   - On the second trip, `name` refers to `"Grace"`.
   - On the third trip, `name` refers to `"Katherine"`.
5. After the last item, the loop ends.

### Trace table

| Trip through loop | `name` value | Printed line |
| --- | --- | --- |
| 1 | `"Ada"` | `Hello, Ada` |
| 2 | `"Grace"` | `Hello, Grace` |
| 3 | `"Katherine"` | `Hello, Katherine` |

A trace table makes an invisible process visible.

### Building a total

Many loops update a value each time:

````python
scores = [10, 8, 9]
total = 0

for score in scores:
    total = total + score

print(total)
````

The variable `total` starts at `0`. Each trip through the loop replaces `total`
with a bigger value.

| Trip | `score` | old `total` | new `total` |
| --- | --- | --- | --- |
| start | — | — | `0` |
| 1 | `10` | `0` | `10` |
| 2 | `8` | `10` | `18` |
| 3 | `9` | `18` | `27` |

### Common mistakes

> **Common mistake**
**Mistake: forgetting to indent the loop body.**

````python
for name in names:
print(name)
````

The `print` line must be indented so Python knows it belongs inside the loop.

> **Common mistake**
**Mistake: resetting the total inside the loop.**

````python
scores = [10, 8, 9]
for score in scores:
    total = 0
    total = total + score
print(total)
````

This prints only the last score because `total` becomes `0` on every trip. Put
`total = 0` before the loop.

## 2. `while` loops repeat while a condition is true

A `while` loop is best when you do not know ahead of time how many repetitions
you need. It keeps running while a condition remains true.

````python
while condition:
    body
````

A safe `while` loop usually has three parts:

1. a starting value;
2. a condition that can eventually become false;
3. an update inside the loop.

### Example 2

````python
count = 3

while count > 0:
    print(count)
    count = count - 1

print("Go!")
````

### Step-by-step explanation

1. `count = 3`
   - The loop starts with `count` equal to `3`.
2. `while count > 0:`
   - Python asks: "Is count greater than 0?"
   - If yes, the indented block runs.
3. `print(count)`
   - The current value is displayed.
4. `count = count - 1`
   - This update is the stopping progress. It moves `count` closer to `0`.
5. When `count` becomes `0`, `count > 0` is false and the loop stops.
6. Python continues after the loop and prints `Go!`.

### Trace table

| Check | `count > 0` | Printed | New `count` |
| --- | --- | --- | --- |
| 1 | `True` | `3` | `2` |
| 2 | `True` | `2` | `1` |
| 3 | `True` | `1` | `0` |
| 4 | `False` | loop stops | — |

### Infinite loops

An **infinite loop** happens when the condition never becomes false.

````python
count = 3
while count > 0:
    print(count)
````

This never changes `count`, so `count > 0` stays true forever. In notebooks,
you may need to interrupt or stop the cell.

## 3. `break` and `continue`

Sometimes a loop needs extra control:

- `break` exits the loop immediately;
- `continue` skips the rest of the current trip and starts the next trip.

Use them sparingly. If you use them, explain why.

### Example 3: Stop when found

````python
names = ["Ada", "Grace", "Katherine"]

for name in names:
    if name == "Grace":
        print("Found Grace!")
        break
````

Python checks names in order. When `name` is `"Grace"`, it prints the message
and `break` stops the loop. `"Katherine"` is never checked.

### Example 4: Skip one item

````python
scores = [10, -1, 8]

for score in scores:
    if score < 0:
        continue
    print(score)
````

The negative score is skipped. The loop still continues with later items.

### Common mistakes

> **Common mistake**
**Mistake: using `break` when you meant `continue`.**

If one bad item should be skipped, use `continue`. If the whole search should
stop, use `break`.

## 4. Tracing loops when they surprise you

When a loop produces an unexpected result, do not guess randomly. Trace it.

### Debugging routine

1. Write down the variables that change.
2. Make a table with one row per loop trip.
3. Fill in values before and after the loop body.
4. Check whether the stopping condition eventually becomes false.
5. Add temporary `print()` calls if the table is hard to fill.

### Practice example

````python
numbers = [2, 4, 6]
total = 0

for number in numbers:
    total = total + number
    print("number:", number, "total:", total)

print("final:", total)
````

Temporary prints are like a flashlight. They help you see what the loop is doing.
After you understand the program, you can remove them.
