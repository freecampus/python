A `with` statement enters a context, runs a block, and then performs cleanup even if an error happens inside the block.

```python
from pathlib import Path

path = Path("hello.txt")
with path.open("w") as file:
file.write("Hello
")

print(path.read_text())
```

### Explain Example 1.1

- `path.open("w")` opens the file for writing.
- `as file` gives the open file a name inside the block.
- The indented write happens while the file is open.
- After the block, the context manager closes the file automatically.

## Common mistakes

> **Common mistake**
Opening files without with can leave resources open if an error interrupts your code.

When you get stuck, use this debugging routine:

1. Read the last line of the error message or inspect the unexpected output.
2. Find the smallest line of code that could be responsible.
3. Print or inspect the value and type at that point.
4. Change one thing.
5. Run again and record what changed.
