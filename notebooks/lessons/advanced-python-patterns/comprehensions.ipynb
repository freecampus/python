## Hands-on episode: Comprehensions

A comprehension builds a new list, dictionary, or set from an iterable by combining an expression, a loop clause, and sometimes a filter clause.

We will learn this by running code, not by memorizing a definition first.
Open the Colab notebook from the button above, find this section, and run each

### Example 1.1

Read the loop version first, then the comprehension as the same idea in a smaller package.

```python
numbers = [1, 2, 3, 4]
squares = [number * number for number in numbers if number % 2 == 0]
print(squares)
```

Run the cell once without editing it. If the result is different from your
prediction, leave the prediction visible and write one sentence about the
difference. That sentence is more useful than a perfect first guess.

### Explain Example 1.1

- `for number in numbers` visits each input value.
- `if number % 2 == 0` keeps only even numbers.
- `number * number` transforms each kept number into its square.
- The brackets collect the transformed values into a new list.

### Debugging checkpoint 1.1

> **Debugging checkpoint**
A comprehension that takes a full minute to read should probably be a regular loop. Readability is more important than cleverness. If debugging is hard, rewrite the comprehension as a loop and print intermediate values.

Do not debug by rewriting the whole example. Read the error type or surprising
output, inspect the closest value with `print(...)` or `type(...)`, then change
one thing. This is the same routine you will use in larger projects.

## Common mistakes

> **Common mistake**
A comprehension that is hard to read should become a regular loop.

When you get stuck, use this debugging routine:

1. Read the last line of the error message or inspect the unexpected output.
2. Find the smallest line of code that could be responsible.
3. Print or inspect the value and type at that point.
4. Change one thing.
5. Run again and record what changed.
