## Hands-on episode: Environments and Dependencies

A dependency is external code your project needs. An environment is the Python interpreter plus installed packages used for one project.

We will learn this by running code, not by memorizing a definition first.

### Debugging checkpoint 1.1

> **Debugging checkpoint**
The most common environment bug is using one Python to install and another Python to run. Always check the interpreter path in VS Code, Jupyter, and the terminal. Recording dependencies is part of the project, not an optional cleanup step.

Do not debug by rewriting the whole example. Read the error type or surprising
output, inspect the closest value with `print(...)` or `type(...)`, then change
one thing. This is the same routine you will use in larger projects.

## How environments connect tools and packages

A virtual environment keeps one project's interpreter and installed packages
separate from other projects.

```{mermaid}

  system["system Python"] -. separate from .-> venv
```

When imports fail, check which Python interpreter is active before reinstalling
packages.

## 1. Virtual Environments

Create isolated Python environments for project dependencies.

> **Analogy**
A virtual environment is a separate lunchbox: one project's ingredients do not spill
into another's.

### What this means

A virtual environment gives one project its own installed packages.

### Example 1
Predict what will happen before you run the code.

```bash
python -m venv .venv
source .venv/bin/activate
python -m pip install requests
```
